```python
import pandas as pd
import matplotlib.pyplot as plt

# Define a function to plot data for a given feature
def plot_data(filename, interval, ax):
    # Read the CSV file
    data = pd.read_csv(f"nisan_data/{filename}{interval}.csv", header=None, names=['Date', 'Value1', 'Value2', 'Value3'], dtype={'Value1': float, 'Value2': float, 'Value3': float})

    # Convert 'Date' column to datetime
    data['Date'] = pd.to_datetime(data['Date'])

    # Plot the data
    for column in data.columns[1:]:
        ax.plot(data['Date'], data[column], label=column)
    ax.set_title(f'{filename} Time Series Data ({interval})')
    ax.set_xlabel('Date')
    ax.set_ylabel('Value')
    ax.legend()

def plot_all_data(files, intervals, figsize=(20, 30)):
    fig, axs = plt.subplots(len(files), len(intervals), figsize=figsize)

    for i, file in enumerate(files):
        for j, interval in enumerate(intervals):
            ax = axs[i, j]
            plot_data(file, interval, ax)

    plt.tight_layout()
    plt.show()

files = ["botTracker", "binanceGlobalAccounts", "binanceTopTraderAccounts", "binanceTopTraderPositions", "sellVolume", "buyVolume", "binanceWhaleDelta", "klines"]
intervals = ["Hourly", "15mins"]

plot_all_data(files, intervals)
```
